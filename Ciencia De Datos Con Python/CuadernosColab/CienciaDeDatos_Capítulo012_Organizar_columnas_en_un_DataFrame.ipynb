## Dataset de trabajo

Para trabajar con organización de columnas vamos a usar una nueva versión del dataset de ventas de una tienda escolar.

Esta vez, el problema principal no estará en los valores de las filas, sino en la estructura de la tabla. Algunas columnas tendrán nombres poco claros, abreviaturas, espacios o estilos de escritura diferentes.

Esto es muy frecuente en datasets reales. A veces recibimos archivos donde las columnas fueron nombradas por distintas personas, exportadas desde sistemas externos o armadas rápidamente sin seguir un criterio uniforme.

Nuestro objetivo será transformar esa estructura inicial en un `DataFrame` más claro y cómodo para analizar.

````python
import pandas as pd

datos = {
    "Producto vendido": [
        "Cuaderno", "Lapicera", "Mochila", "Regla", "Cartuchera",
        "Calculadora", "Auriculares", "Resaltadores", "Guardapolvo", "Compas"
    ],
    "Cat": [
        "Libreria", "Libreria", "Accesorios", "Libreria", "Accesorios",
        "Tecnologia", "Tecnologia", "Libreria", "Indumentaria", "Libreria"
    ],
    "Precio Unitario": [
        1200, 350, 8500, 500, 3200,
        18500, 7600, 2100, 14500, 2600
    ],
    "Cant.": [
        15, 40, 5, 25, 8,
        3, 6, 12, 4, 10
    ],
    "Sucursal Venta": [
        "Centro", "Centro", "Norte", "Sur", "Norte",
        "Centro", "Sur", "Centro", "Norte", "Sur"
    ],
    "Pago": [
        "Efectivo", "Debito", "Credito", "Efectivo", "Debito",
        "Credito", "Credito", "Debito", "Credito", "Efectivo"
    ],
    "Fecha venta": [
        "2024-03-01", "2024-03-01", "2024-03-02", "2024-03-02", "2024-03-03",
        "2024-03-03", "2024-03-04", "2024-03-04", "2024-03-05", "2024-03-05"
    ],
    "Código interno": [
        "A001", "A002", "B010", "A003", "B011",
        "T020", "T021", "A004", "I030", "A005"
    ]
}

df = pd.DataFrame(datos)

df
````

Al observar el `DataFrame`, vemos que los datos son comprensibles, pero los nombres de las columnas no siguen un criterio uniforme.

Algunas columnas tienen espacios, como `Producto vendido` o `Precio Unitario`. Otras usan abreviaturas, como `Cat` o `Cant.`. También hay nombres que podrían ser más descriptivos, como `Pago`, que en realidad representa la forma de pago.

Estos detalles pueden parecer menores, pero afectan la claridad del código. Es más cómodo y menos propenso a errores trabajar con nombres de columnas consistentes, descriptivos y fáciles de escribir.


## Qué significa organizar columnas

Organizar columnas significa revisar y ajustar la estructura del `DataFrame` para que sea más clara y fácil de usar.

Esto puede incluir varias tareas. Por ejemplo, podemos renombrar columnas para que tengan nombres más descriptivos, eliminar columnas que no necesitamos, reordenar variables para que la tabla sea más legible o construir una versión final del dataset con las columnas que vamos a usar en el análisis.

En este capítulo no nos vamos a concentrar en corregir los valores dentro de las columnas, sino en la estructura de la tabla.

Por ejemplo, una columna llamada `Cat` puede entenderse si conocemos el contexto, pero no es tan clara como `categoria`. Una columna llamada `Cant.` puede ser breve, pero un nombre como `cantidad_vendida` describe mejor qué representa la variable.

También es importante evitar nombres incómodos para escribir código. Los nombres con espacios, puntos o mayúsculas mezcladas pueden funcionar, pero suelen ser menos prácticos. Por ejemplo:

```python
df["Producto vendido"]
```

funciona correctamente, pero es más cómodo trabajar con:

```python
df["producto"]
```

o con:

```python
df["producto_vendido"]
```

Una buena organización de columnas no cambia los datos en sí mismos, pero mejora la claridad del trabajo. Hace que el código sea más fácil de leer, reduce errores de escritura y ayuda a que otras personas comprendan mejor el dataset.

## Renombrar columnas con rename()

Una de las formas más claras de cambiar nombres de columnas en Pandas es usar el método `rename()`.

Este método permite indicar qué columnas queremos renombrar y qué nuevo nombre debe recibir cada una.

Para eso usamos un diccionario. En ese diccionario, cada clave es el nombre actual de una columna, y cada valor es el nuevo nombre que queremos asignarle.

````python
df = df.rename(columns={
    "Producto vendido": "producto",
    "Cat": "categoria",
    "Precio Unitario": "precio_unitario",
    "Cant.": "cantidad_vendida",
    "Sucursal Venta": "sucursal",
    "Pago": "forma_pago",
    "Fecha venta": "fecha"
})

df
````

Ahora el `DataFrame` tiene nombres de columnas más claros y consistentes.

Podemos revisar nuevamente los nombres:

````python
df.columns
````

Los nombres nuevos siguen un criterio común:

```text
producto
categoria
precio_unitario
cantidad_vendida
sucursal
forma_pago
fecha
Código interno
```

La columna `Código interno` todavía conserva su nombre original. Esto ocurrió porque no la incluimos en el diccionario de reemplazos.

Con `rename()`, Pandas solo cambia los nombres de las columnas indicadas. Las columnas que no aparecen en el diccionario quedan como estaban.

Esto puede ser útil cuando queremos modificar solo algunas columnas sin afectar el resto del `DataFrame`.

## Guardar el resultado de rename()

En la celda anterior escribimos:

```python
df = df.rename(columns={
    "Producto vendido": "producto",
    "Cat": "categoria",
    "Precio Unitario": "precio_unitario",
    "Cant.": "cantidad_vendida",
    "Sucursal Venta": "sucursal",
    "Pago": "forma_pago",
    "Fecha venta": "fecha"
})
```

Es importante observar que asignamos el resultado nuevamente a la variable `df`.

Esto significa que reemplazamos el `DataFrame` anterior por una nueva versión con los nombres de columnas modificados.

Si usáramos `rename()` sin guardar el resultado, Pandas mostraría una versión renombrada, pero el `DataFrame` original seguiría igual.

Veamos un ejemplo pequeño para entenderlo mejor.

````python
df_ejemplo = pd.DataFrame({
    "Nombre Producto": ["Cuaderno", "Lapicera"],
    "Precio": [1200, 350]
})

df_ejemplo
````

````python
df_ejemplo.rename(columns={
    "Nombre Producto": "producto",
    "Precio": "precio"
})
````

La salida muestra una versión con nombres modificados. Sin embargo, si volvemos a mostrar `df_ejemplo`, veremos que conserva los nombres originales.

````python
df_ejemplo
````

Para conservar el cambio, debemos guardar el resultado:

````python
df_ejemplo = df_ejemplo.rename(columns={
    "Nombre Producto": "producto",
    "Precio": "precio"
})

df_ejemplo
````

Esta idea aparece muchas veces en Pandas. Muchos métodos devuelven un nuevo resultado, pero no modifican automáticamente el objeto original. Por eso, cuando queremos conservar una transformación, debemos asignarla a una variable.

## Eliminar columnas con drop()

Además de renombrar columnas, a veces necesitamos eliminar columnas que no vamos a utilizar.

En nuestro `DataFrame` hay una columna llamada `Código interno`. Esa columna podría ser útil para un sistema de gestión, pero tal vez no sea necesaria para el análisis que queremos realizar.

Si decidimos quitarla, podemos usar el método `drop()`.

Para eliminar columnas, usamos el parámetro `columns`.

````python
df = df.drop(columns=["Código interno"])

df
````

Ahora el `DataFrame` ya no contiene la columna `Código interno`.

Al igual que ocurría con `rename()`, usamos una asignación:

```python
df = df.drop(columns=["Código interno"])
```

Esto significa que guardamos en `df` una nueva versión del `DataFrame`, pero sin esa columna.

Eliminar columnas puede ser útil cuando una variable no aporta información para el análisis, cuando contiene datos redundantes o cuando queremos construir una versión más simple del dataset. Sin embargo, conviene ser cuidadosos. Una columna que parece innecesaria al comienzo puede servir más adelante para verificar información, cruzar datos o reconstruir decisiones.

Por eso, antes de eliminar una columna conviene preguntarnos:

```text
¿Esta columna aporta información útil para el análisis?
¿La información aparece repetida en otra columna?
¿Podría necesitarla más adelante para verificar o cruzar datos?
¿Estoy trabajando sobre una copia o sobre el único DataFrame disponible?
```

En general, si no estamos seguros, puede ser mejor conservar la columna o crear una versión organizada del `DataFrame` en una nueva variable.

## Eliminar más de una columna

`drop()` también permite eliminar varias columnas al mismo tiempo.

Para hacerlo, pasamos una lista con los nombres de las columnas que queremos quitar.

Veamos un ejemplo con un `DataFrame` auxiliar. Vamos a recrear una versión pequeña del dataset para mostrar la eliminación de varias columnas sin afectar el `DataFrame` principal que ya venimos organizando.

````python
df_auxiliar = pd.DataFrame(datos)

df_auxiliar.head()
````

En este `DataFrame` auxiliar todavía aparecen los nombres originales de las columnas.

Supongamos que queremos eliminar las columnas `Código interno` y `Fecha venta`.

````python
df_auxiliar = df_auxiliar.drop(columns=["Código interno", "Fecha venta"])

df_auxiliar.head()
````

Ahora `df_auxiliar` ya no contiene esas dos columnas.

La estructura general es:

```python
df.drop(columns=["columna_1", "columna_2"])
```

Eliminar varias columnas puede ser útil cuando queremos simplificar un dataset y quedarnos solamente con las variables necesarias. Sin embargo, conviene hacerlo con cuidado. Si eliminamos columnas demasiado pronto, podríamos perder información que luego necesitábamos para revisar el origen de los datos, controlar errores o enriquecer el análisis.

Por eso, muchas veces es preferible trabajar con una copia o con una nueva variable, especialmente durante las primeras etapas de exploración y preparación.

## Reordenar columnas

Además de renombrar o eliminar columnas, también podemos cambiar el orden en que aparecen dentro del `DataFrame`.

Reordenar columnas no modifica los valores. Solo cambia la forma en que se presenta la tabla.

Esto puede ser útil cuando queremos que las variables sigan un orden más lógico. Por ejemplo, podríamos preferir que primero aparezcan las columnas que identifican la venta, luego las variables numéricas y finalmente la información contextual.

En nuestro `DataFrame` actual tenemos estas columnas:

````python
df.columns
````

Podemos crear una lista con el orden deseado:

````python
columnas_ordenadas = [
    "producto",
    "categoria",
    "precio_unitario",
    "cantidad_vendida",
    "sucursal",
    "forma_pago",
    "fecha"
]
````

Luego usamos esa lista para reorganizar el `DataFrame`:

````python
df = df[columnas_ordenadas]

df
````

Ahora las columnas aparecen en el orden que indicamos en la lista.

Esta técnica es muy útil porque nos permite construir una versión final del dataset con una estructura más clara.

La idea general es:

```python
df = df[lista_de_columnas_en_el_orden_deseado]
```

Es importante que todos los nombres de la lista existan en el `DataFrame`. Si escribimos mal un nombre de columna o incluimos una columna que no existe, Pandas mostrará un error.

## Construir una versión organizada del DataFrame

Hasta ahora aplicamos varias operaciones por separado: renombramos columnas, eliminamos una columna que no necesitábamos y reordenamos las variables.

En un trabajo real, muchas veces queremos construir una versión organizada del dataset para usarla en el análisis posterior. Esa versión puede conservar solamente las columnas necesarias, con nombres claros y en un orden lógico.

Una forma de hacerlo es crear una nueva variable, por ejemplo `df_organizado`.

Para mostrar el proceso completo, vamos a recrear el `DataFrame` original y aplicar las transformaciones principales.

````python
df_original = pd.DataFrame(datos)

df_organizado = df_original.rename(columns={
    "Producto vendido": "producto",
    "Cat": "categoria",
    "Precio Unitario": "precio_unitario",
    "Cant.": "cantidad_vendida",
    "Sucursal Venta": "sucursal",
    "Pago": "forma_pago",
    "Fecha venta": "fecha"
})

df_organizado = df_organizado.drop(columns=["Código interno"])

df_organizado = df_organizado[
    [
        "producto",
        "categoria",
        "precio_unitario",
        "cantidad_vendida",
        "sucursal",
        "forma_pago",
        "fecha"
    ]
]

df_organizado
````

En este bloque realizamos tres pasos.

Primero renombramos las columnas para que tengan nombres más claros y consistentes. Luego eliminamos la columna `Código interno`, porque decidimos que no era necesaria para este análisis. Finalmente reordenamos las columnas para que la tabla quede presentada de una manera más cómoda.

El `DataFrame` original sigue estando disponible en la variable `df_original`, mientras que la versión organizada quedó guardada en `df_organizado`.

Esta forma de trabajo es recomendable porque conserva una versión inicial de los datos y, al mismo tiempo, nos permite construir una versión más limpia y clara para continuar el análisis.

## Verificar la estructura final

Después de organizar las columnas, conviene revisar cómo quedó el `DataFrame`.

Podemos empezar consultando sus nombres de columnas:

````python
df_organizado.columns
````

También podemos revisar sus dimensiones:

````python
df_organizado.shape
````

Y podemos observar las primeras filas:

````python
df_organizado.head()
````

Estas verificaciones nos permiten confirmar que la organización se aplicó correctamente.

El `DataFrame` organizado debería tener nombres de columnas claros, no debería incluir columnas que decidimos eliminar y debería presentar las variables en un orden lógico.

También podemos comparar las columnas originales con las columnas organizadas:

````python
print("Columnas originales:")
print(list(df_original.columns))

print()

print("Columnas organizadas:")
print(list(df_organizado.columns))
````

Esta comparación muestra con claridad qué cambió.

El dataset original conservaba nombres como `Producto vendido`, `Cat`, `Precio Unitario`, `Cant.`, `Sucursal Venta`, `Pago`, `Fecha venta` y `Código interno`.

La versión organizada, en cambio, tiene nombres más consistentes y ya no incluye la columna que decidimos quitar.

Verificar la estructura final es una buena práctica. Cada vez que renombramos, eliminamos o reordenamos columnas, conviene revisar que el resultado coincida con lo que esperábamos.

## Errores frecuentes al organizar columnas

Al organizar columnas en Pandas pueden aparecer algunos errores comunes.

Uno de los más frecuentes es escribir mal el nombre de una columna. Esto puede ocurrir al renombrar, eliminar o reordenar.

Por ejemplo, si intentamos eliminar una columna llamada `"Codigo interno"` sin tilde, pero en el `DataFrame` la columna se llama `"Código interno"` con tilde, Pandas no encontrará esa columna.

````python
# Esta instrucción produciría un error porque la columna se llama "Código interno".
# df_original.drop(columns=["Codigo interno"])
````

Para evitar este tipo de problemas, conviene revisar los nombres exactos con:

````python
df_original.columns
````

Otro error frecuente es olvidar guardar el resultado de una operación. Métodos como `rename()` y `drop()` devuelven una nueva versión del `DataFrame`, pero no modifican automáticamente el objeto original si no guardamos el resultado.

Por ejemplo:

````python
df_ejemplo = pd.DataFrame({
    "Producto vendido": ["Cuaderno", "Lapicera"],
    "Precio Unitario": [1200, 350]
})

df_ejemplo.rename(columns={
    "Producto vendido": "producto",
    "Precio Unitario": "precio_unitario"
})
````

La salida parece mostrar columnas renombradas, pero si volvemos a revisar `df_ejemplo`, veremos que conserva los nombres originales:

````python
df_ejemplo
````

Para conservar el cambio, debemos guardar el resultado:

````python
df_ejemplo = df_ejemplo.rename(columns={
    "Producto vendido": "producto",
    "Precio Unitario": "precio_unitario"
})

df_ejemplo
````

También puede aparecer un error al reordenar columnas si la lista contiene un nombre que no existe o si olvidamos incluir una columna que queríamos conservar.

````python
# Esta instrucción produciría un error porque "precio" no existe en df_organizado.
# La columna se llama "precio_unitario".

# df_organizado[["producto", "categoria", "precio"]]
````

Finalmente, debemos tener cuidado al eliminar columnas demasiado pronto. Eliminar una columna puede simplificar el dataset, pero también puede hacernos perder información útil para controles posteriores.

Una buena rutina para organizar columnas podría ser:

```text
revisar nombres actuales
renombrar columnas necesarias
decidir qué columnas conservar o eliminar
reordenar variables
verificar la estructura final
```

Organizar columnas no es solo una cuestión estética. Una estructura clara ayuda a escribir código más legible, reduce errores y facilita el análisis posterior.
