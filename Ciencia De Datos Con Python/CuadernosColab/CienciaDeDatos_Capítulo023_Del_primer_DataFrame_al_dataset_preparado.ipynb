## De mirar una tabla a comprender un DataFrame

El recorrido comenzó con una idea básica: un `DataFrame` es una estructura de datos organizada en filas y columnas.

A simple vista puede parecer una tabla común, parecida a una hoja de cálculo. Sin embargo, en Pandas un `DataFrame` es mucho más que una tabla visible. Es una estructura que permite seleccionar datos, filtrarlos, ordenarlos, transformarlos, combinarlos, resumirlos y analizarlos mediante código.

Por eso, los primeros pasos fueron importantes. Antes de modificar un dataset, necesitamos aprender a observarlo.

Herramientas como `head()`, `tail()`, `shape`, `columns` e `info()` nos ayudaron a responder preguntas iniciales:

```text
¿Cuántas filas y columnas tiene el dataset?
¿Cómo se llaman sus columnas?
Qué tipos de datos interpretó Pandas?
¿Hay valores faltantes?
¿Cómo se ven los primeros registros?
```

Estas preguntas pueden parecer simples, pero forman parte de una práctica fundamental: no trabajar a ciegas.

Cuando recibimos un dataset, la primera tarea no es aplicar transformaciones ni calcular resultados. La primera tarea es entender qué tenemos delante.

Esta idea se mantuvo durante toda la etapa. Cada vez que avanzamos hacia una transformación más compleja, volvimos a la misma lógica: observar primero, decidir después.

## Seleccionar, filtrar y ordenar información

Después de aprender a inspeccionar un `DataFrame`, el siguiente paso fue aprender a trabajar con partes específicas de la tabla.

No siempre necesitamos usar todos los datos al mismo tiempo. Muchas preguntas se responden seleccionando ciertas columnas, ciertas filas o ciertos registros que cumplen una condición.

Por eso trabajamos con selección de columnas, selección de filas, uso de etiquetas con `loc`, filtros con condiciones y combinación de condiciones.

Ese bloque del recorrido nos permitió empezar a formular preguntas más concretas:

```text
¿Qué columnas necesito para responder esta pregunta?
¿Qué filas cumplen cierta condición?
¿Qué registros pertenecen a una categoría específica?
¿Qué ventas superan determinado valor?
¿Qué casos cumplen varias condiciones al mismo tiempo?
```

Aprender a filtrar fue un paso muy importante, porque los filtros transforman una tabla general en una respuesta más precisa. En lugar de mirar todo el dataset, empezamos a recortar la información según un criterio.

Luego incorporamos el ordenamiento. Ordenar datos nos permitió observar los registros desde otra perspectiva: productos más caros, ventas con mayor cantidad, valores más bajos, primeros y últimos registros según algún criterio.

A partir de ahí aparecieron los primeros rankings y la búsqueda de valores extremos. Eso nos ayudó a entender que el orden de una tabla no es neutro: según la columna elegida para ordenar, podemos responder preguntas distintas.

Una idea importante de esta parte fue que seleccionar, filtrar y ordenar no son operaciones aisladas. Muchas veces se combinan. Podemos filtrar un subconjunto y luego ordenarlo, seleccionar algunas columnas para leer mejor el resultado, o construir un ranking solo dentro de una categoría.

Ese fue un paso importante: empezamos a usar Pandas no solo para mostrar datos, sino para interrogarlos.

## Crear, modificar y organizar columnas

Una vez que aprendimos a seleccionar, filtrar y ordenar datos, dimos otro paso importante: empezamos a transformar la estructura del `DataFrame`.

Crear nuevas columnas nos permitió construir información que no estaba explícita en el dataset original. Por ejemplo, si tenemos una cantidad y un precio unitario, podemos calcular un importe total. Si tenemos una fecha, podemos extraer el año, el mes o el día de la semana. Si tenemos una condición, podemos crear una nueva columna que clasifique los registros según algún criterio.

Este tipo de transformación es muy importante porque muchos datasets no vienen con todas las variables que necesitamos. A veces los datos disponibles son solo el punto de partida. El análisis requiere construir nuevas variables que representen mejor la pregunta que queremos responder.

También aprendimos a modificar columnas existentes. Eso nos permitió corregir textos, reemplazar valores, ajustar formatos y decidir si convenía sobrescribir una columna o crear una versión nueva. Esa decisión no es menor: conservar una columna original puede ser útil para comparar antes y después de una transformación.

Luego trabajamos con la organización general del `DataFrame`: renombrar columnas, eliminar columnas que no necesitamos y reordenar variables. Estas acciones no siempre cambian los datos en sí mismos, pero hacen que la tabla sea más clara, más cómoda de usar y más fácil de interpretar.

La idea central de esta parte fue que un `DataFrame` no es una estructura fija. Podemos transformarlo para que represente mejor nuestro problema de análisis.

Pero esa transformación debe ser cuidadosa. Cada columna nueva o modificada debería tener un propósito: facilitar una pregunta, corregir un problema, validar una relación o preparar el dataset para una etapa posterior.

## De operaciones aisladas a un proceso de limpieza

En la primera parte del recorrido aprendimos muchas operaciones importantes de Pandas. Cada una tenía un objetivo relativamente claro: seleccionar columnas, filtrar filas, ordenar datos, crear una columna, renombrar variables o revisar la estructura del `DataFrame`.

Pero, cuando empezamos a trabajar con limpieza de datos, el enfoque cambió.

Ya no se trataba solamente de aprender una instrucción nueva, sino de pensar un proceso completo. Un dataset real puede tener varios problemas al mismo tiempo: valores faltantes, duplicados, textos inconsistentes, columnas con tipos incorrectos, fechas mal interpretadas o valores que no tienen sentido dentro del contexto.

Por eso, la limpieza de datos nos obligó a trabajar con una lógica más amplia:

```text
diagnosticar
decidir
transformar
verificar
```

Diagnosticar significa observar el problema antes de actuar. Decidir implica elegir una estrategia razonable según el contexto. Transformar consiste en aplicar una modificación concreta. Verificar significa comprobar que el cambio produjo el resultado esperado y no generó nuevos problemas.

Esta forma de trabajo fue apareciendo en distintos temas. Con los valores faltantes, primero los contamos y analizamos antes de eliminarlos o imputarlos. Con los duplicados, revisamos si las filas eran realmente repeticiones o si podían representar casos válidos. Con las categorías, verificamos si las diferencias eran solo de escritura o si implicaban significados distintos. Con los tipos de datos, convertimos columnas pero luego revisamos cuántos valores se transformaron en `NaN` o `NaT`.

Ese fue uno de los aprendizajes más importantes de la etapa: limpiar datos no es aplicar comandos automáticamente. Es tomar decisiones justificadas y dejar evidencia de esas decisiones.

## Qué problemas aprendimos a detectar y tratar

Durante la etapa de limpieza aparecieron varios problemas frecuentes en el trabajo con datos.

El primero fue el de los valores faltantes. Aprendimos que un valor ausente no siempre significa lo mismo. Puede tratarse de un dato que no fue registrado, de un campo que no correspondía, de un error de carga o de un valor problemático que todavía no fue interpretado como faltante. Por eso, antes de eliminar o completar datos, analizamos cuántos faltantes había, qué porcentaje representaban y qué columnas estaban afectadas.

Luego trabajamos con estrategias para tratar esos faltantes. Vimos que eliminar filas puede ser razonable en algunos casos, pero también puede reducir el dataset o sesgar el análisis. Imputar permite conservar registros, pero introduce valores que no estaban originalmente en los datos. Reconstruir valores a partir de otras columnas puede ser una alternativa más sólida, siempre que exista una relación confiable.

También revisamos duplicados. Allí aprendimos que no todas las filas parecidas son necesariamente registros repetidos. En algunos contextos, dos filas iguales pueden indicar una carga duplicada. En otros, dos filas similares pueden representar eventos reales distintos. La clave fue definir el criterio de duplicación antes de eliminar registros.

Más adelante trabajamos con textos y categorías. Vimos que diferencias pequeñas, como espacios, mayúsculas, minúsculas, tildes o variantes de escritura, pueden fragmentar los conteos. Para una persona, dos valores pueden parecer equivalentes; para Pandas, si no están escritos exactamente igual, son valores distintos. Por eso limpiamos textos, unificamos categorías y revisamos valores problemáticos como `UNKNOWN` o `ERROR`.

Después avanzamos sobre tipos de datos. Aprendimos que una columna puede parecer numérica pero estar cargada como texto. También vimos que una fecha puede verse como fecha, pero seguir siendo un `object` para Pandas. Convertir tipos con cuidado fue fundamental para poder calcular, ordenar, comparar y analizar correctamente.

Finalmente trabajamos con fechas y datos temporales. Allí aprendimos que las fechas permiten crear nuevas variables, revisar rangos temporales, ordenar cronológicamente y comenzar a analizar datos por períodos.

Todos estos problemas tienen algo en común: no se resuelven solo con una instrucción. Requieren interpretar qué representan dentro del dataset y qué impacto pueden tener sobre el análisis.

## Qué significa tener un dataset preparado

Después de todo este recorrido, podemos preguntarnos qué significa realmente tener un dataset preparado.

Un dataset preparado no es necesariamente un dataset perfecto. En muchos casos, la perfección no existe: puede haber información faltante que no podemos recuperar, valores desconocidos que deben conservarse, registros que requieren revisión externa o decisiones que dependen del objetivo del análisis.

Preparar un dataset significa otra cosa.

Significa que entendemos mejor qué contiene la tabla, qué problemas tenía, qué transformaciones aplicamos y qué limitaciones siguen presentes.

Un dataset preparado debería tener, al menos, algunas características:

```text
nombres de columnas claros
tipos de datos adecuados
valores problemáticos identificados
duplicados revisados
categorías más consistentes
fechas interpretadas correctamente
columnas útiles para el análisis
validaciones básicas realizadas
problemas pendientes documentados
```

Esto no garantiza que todas las conclusiones futuras sean correctas. Pero reduce el riesgo de trabajar sobre datos mal interpretados.

Por ejemplo, si una columna de importes estaba cargada como texto y no lo advertimos, podríamos calcular mal o no poder calcular. Si una categoría aparece escrita de cinco formas distintas, podríamos subestimar su frecuencia real. Si una fecha no fue convertida correctamente, podríamos excluir registros del análisis temporal sin darnos cuenta.

La preparación de datos busca evitar ese tipo de errores.

Por eso, preparar datos no es una etapa secundaria. Es una parte central del análisis. Antes de buscar patrones, construir gráficos o sacar conclusiones, necesitamos asegurarnos de que el dataset esté en condiciones razonables para responder las preguntas que queremos formular.

## Buenas prácticas que conviene conservar

A partir de lo trabajado hasta aquí, podemos reconocer algunas buenas prácticas que conviene mantener en los próximos capítulos.

La primera es inspeccionar antes de transformar. Cada vez que recibimos un dataset, necesitamos observar su estructura, sus columnas, sus tipos de datos y sus primeras filas. Esa inspección inicial evita trabajar a ciegas y permite detectar señales tempranas de problemas.

La segunda es no aplicar soluciones automáticas sin contexto. Métodos como `dropna()`, `fillna()`, `replace()` o `drop_duplicates()` son útiles, pero no deberían usarse sin pensar qué representan los datos afectados. Una instrucción puede ser correcta desde el punto de vista técnico y, al mismo tiempo, ser una mala decisión para el análisis.

La tercera es trabajar con copias cuando vamos a realizar transformaciones importantes. Conservar una versión original o una copia de referencia permite comparar antes y después, revisar decisiones y volver atrás si una transformación no resulta adecuada.

La cuarta es verificar después de cada cambio relevante. Si convertimos tipos, revisamos `dtypes` o `info()`. Si limpiamos categorías, revisamos `value_counts()`. Si eliminamos duplicados, contamos nuevamente los duplicados. Si creamos columnas nuevas, observamos ejemplos y comprobamos que tengan sentido.

La quinta es documentar las decisiones. En limpieza de datos, muchas acciones dependen del criterio adoptado. Por ejemplo, decidir que `"ERROR"` se agrupe dentro de `"UNKNOWN"` puede ser razonable en un contexto, pero no en otro. Por eso, conviene dejar claro qué se hizo y por qué.

La sexta es distinguir entre datos corregidos y problemas pendientes. Un dataset puede quedar preparado aunque todavía conserve faltantes o valores desconocidos. Lo importante es saber que esos casos existen, cómo están representados y qué limitaciones generan para el análisis.

Estas prácticas no pertenecen solo a la etapa de limpieza. Van a seguir siendo útiles cuando avancemos hacia resúmenes, agrupamientos, visualizaciones e interpretación de resultados.
