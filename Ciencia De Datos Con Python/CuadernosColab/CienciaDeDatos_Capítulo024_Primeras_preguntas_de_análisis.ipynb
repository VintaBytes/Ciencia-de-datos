## Usar un dataset de práctica

Para comenzar esta nueva etapa vamos a usar un dataset de práctica incluido en una librería de Python.

Muchas librerías orientadas al análisis de datos, la visualización o el aprendizaje automático incluyen datasets pequeños y conocidos. Estos datasets se usan para aprender, probar herramientas, comparar resultados y practicar sin depender siempre de archivos externos.

Esto tiene una ventaja importante: podemos concentrarnos en las preguntas de análisis sin comenzar otra vez por la descarga, la carga manual o la limpieza profunda de los datos.

En este caso vamos a usar el dataset `tips`, disponible desde la librería Seaborn.

Este dataset contiene información sobre cuentas y propinas en un restaurante. Cada fila representa una mesa atendida, e incluye datos como el total de la cuenta, la propina, el día, el horario y la cantidad de personas.

En esta etapa no vamos a limpiar el dataset. Lo vamos a usar como punto de partida para formular y responder primeras preguntas de análisis.

En otras palabras, el dataset ya nos ofrece dos caminos iniciales de análisis: resumir valores numéricos y contar categorías.

- ¿Cuál es el tamaño de mesa más habitual?

Estas preguntas todavía son simples, pero marcan un cambio importante: ya no estamos revisando el dataset para prepararlo, sino para entender qué información contiene.

## Contar valores en una columna categórica

Hasta ahora observamos resúmenes generales de las columnas numéricas. Ahora vamos a mirar una columna categórica.

Una primera pregunta posible es:

**¿Qué días aparecen registrados en el dataset y cuántas cuentas hay en cada uno?**

Para responderla podemos contar cuántas veces aparece cada valor en la columna `day`.

```python
df["day"].value_counts()
```

La columna `day` muestra registros para cuatro días: `Thur`, `Fri`, `Sat` y `Sun`.

El día con más cuentas registradas es **sábado**, con **87 cuentas**. Luego aparecen **domingo**, con **76**, y **jueves**, con **62**. El día con menos registros es **viernes**, con **19 cuentas**.

Este resultado no significa necesariamente que el sábado sea “mejor” o “más rentable”. Por ahora solo sabemos que en el dataset hay más cuentas registradas ese día. Para hablar de ingresos, promedios o comportamiento de consumo necesitaríamos hacer otros cálculos.

## Contar registros por horario

También podemos preguntarnos en qué momento del día se registraron más cuentas.

La columna `time` indica si cada registro corresponde al almuerzo o a la cena. Contar sus valores nos permite ver cómo se distribuyen las cuentas entre esos dos momentos.

```python
df["time"].value_counts()
```

La mayor parte de las cuentas registradas corresponde a **Dinner**, con **176 registros**. En cambio, **Lunch** aparece con **68 registros**.

Esto nos permite ver que el dataset tiene más observaciones de cenas que de almuerzos.

Esta salida solo describe la cantidad de registros por horario. Para analizar consumo, propinas o tamaño de mesa necesitaremos combinar esta información con columnas numéricas como `total_bill`, `tip` o `size`.

## Analizar el tamaño de las mesas

Otra pregunta inicial es cuántas personas había en cada mesa.

La columna `size` indica el tamaño de la mesa. Aunque es una columna numérica, en este caso también puede analizarse como una categoría: mesa de 1 persona, de 2 personas, de 3 personas, y así sucesivamente.

Contar estos valores nos permite saber qué tamaños de mesa aparecen con más frecuencia en el dataset.

```python
df["size"].value_counts()
```

El tamaño de mesa más frecuente es **2 personas**, con **156 registros**.

Luego aparecen las mesas de **3 personas** y **4 personas**, con cantidades bastante similares: **38** y **37 registros** respectivamente.

Los tamaños menos frecuentes son las mesas de **1**, **5** y **6** personas.

Este resultado nos muestra que el dataset está dominado por mesas pequeñas, especialmente de dos personas. Esta información puede ser importante más adelante, porque el tamaño de la mesa puede influir en el total de la cuenta y también en la propina.

## Calcular indicadores simples

Además de contar registros, podemos calcular algunos indicadores numéricos básicos.

Por ejemplo, podemos preguntarnos:

- ¿Cuál fue el total facturado en todas las cuentas?
- ¿Cuánto se dejó de propina en total?
- ¿Cuál fue el promedio del total de cuenta?
- ¿Cuál fue el promedio de propina?

Estas preguntas todavía no comparan grupos ni categorías. Son indicadores generales del dataset completo.

```python
total_facturado = df["total_bill"].sum()
total_propinas = df["tip"].sum()
promedio_cuenta = df["total_bill"].mean()
promedio_propina = df["tip"].mean()

total_facturado, total_propinas, promedio_cuenta, promedio_propina
```

Los resultados obtenidos corresponden a cuatro indicadores generales del dataset:

- El total facturado es **4827.77**.
- El total de propinas es **731.58**.
- El promedio de cuenta es aproximadamente **19.79**.
- El promedio de propina es aproximadamente **3.00**.

Estos valores resumen el conjunto completo de datos, sin distinguir todavía por día, horario, tamaño de mesa u otras categorías.

La salida es correcta, pero no es muy cómoda para leer: aparece como una tupla de valores y Pandas/NumPy muestra algunos números con demasiados decimales. Por eso conviene construir una tabla pequeña de indicadores.

## Presentar indicadores en una tabla

Cuando calculamos varios indicadores, puede ser más claro organizarlos en una pequeña tabla.

Esto no cambia los resultados, pero mejora la lectura y acerca el análisis a una forma más comunicable.

```python
indicadores_generales = pd.DataFrame({
    "indicador": [
        "Total facturado",
        "Total de propinas",
        "Promedio de cuenta",
        "Promedio de propina"
    ],
    "valor": [
        total_facturado,
        total_propinas,
        promedio_cuenta,
        promedio_propina
    ]
})

indicadores_generales
```

Ahora los indicadores aparecen organizados en una tabla.

Esta forma de presentación es más clara que una salida con varios valores sueltos, porque cada número queda asociado a su significado.

Sin embargo, los valores todavía aparecen con muchos decimales. Para comunicar mejor estos resultados, podemos redondearlos a dos decimales.

## Redondear resultados

Cuando los indicadores tienen muchos decimales, la lectura puede volverse innecesariamente pesada.

En este caso, redondear a dos decimales alcanza para comunicar los resultados de manera clara.

```python
indicadores_generales["valor"] = indicadores_generales["valor"].round(2)

indicadores_generales
```

La tabla final presenta los indicadores generales de manera más clara.

Podemos resumir esta primera lectura así:

- En el dataset se registró un total facturado de **4827.77**.
- El total de propinas fue de **731.58**.
- El promedio de cuenta fue de **19.79**.
- El promedio de propina fue de **3.00**.

Estos indicadores permiten tener una primera idea general del dataset. Sin embargo, todavía no muestran diferencias internas.

Por ejemplo, no sabemos si el promedio de cuenta cambia entre almuerzo y cena, si las propinas varían según el día o si las mesas más grandes generan cuentas más altas. Para responder ese tipo de preguntas vamos a necesitar comparar grupos.

## Comparar una pregunta en dos subconjuntos

Hasta ahora calculamos indicadores para todo el dataset.

Pero una pregunta de análisis muchas veces necesita comparar partes del conjunto de datos. Por ejemplo:

**¿El promedio de cuenta es igual en el almuerzo y en la cena?**

Todavía no necesitamos una herramienta nueva para responder esto. Podemos usar filtros, que ya conocemos, para separar las cuentas de almuerzo y cena, y luego calcular el promedio de `total_bill` en cada caso.

```python
promedio_almuerzo = df[df["time"] == "Lunch"]["total_bill"].mean()
promedio_cena = df[df["time"] == "Dinner"]["total_bill"].mean()

promedio_almuerzo, promedio_cena
```

El promedio de cuenta durante el almuerzo es aproximadamente **17.17**.

El promedio de cuenta durante la cena es aproximadamente **20.80**.

En este primer cálculo, las cuentas de la cena tienen un promedio más alto que las cuentas del almuerzo.

De todos modos, la salida vuelve a aparecer como dos valores sueltos. Para comparar mejor, conviene organizar el resultado en una pequeña tabla.

## Organizar la comparación

Cuando comparamos dos valores, una tabla simple puede ayudar a leer mejor el resultado.

Vamos a construir un pequeño `DataFrame` con el promedio de cuenta para cada horario.

```python
comparacion_horario = pd.DataFrame({
    "horario": ["Lunch", "Dinner"],
    "promedio_total_cuenta": [promedio_almuerzo, promedio_cena]
})

comparacion_horario["promedio_total_cuenta"] = comparacion_horario["promedio_total_cuenta"].round(2)

comparacion_horario
```

La tabla permite comparar con más claridad el promedio de cuenta según el horario.

En este dataset, las cuentas registradas durante la cena (`Dinner`) tienen un promedio de **20.80**, mientras que las cuentas registradas durante el almuerzo (`Lunch`) tienen un promedio de **17.17**.

Esto sugiere que, en estos datos, el consumo promedio es mayor durante la cena.

Esta comparación se resolvió con filtros, una herramienta que ya conocemos. Más adelante vamos a estudiar formas más directas de resumir datos por grupos, especialmente con `groupby()`.

## Una respuesta puede abrir nuevas preguntas

Un análisis exploratorio no avanza en línea recta. Muchas veces, una respuesta genera nuevas preguntas.

Por ejemplo, acabamos de observar que el promedio de cuenta es mayor durante la cena que durante el almuerzo. A partir de eso podríamos preguntarnos:

- ¿Hay muchas más cuentas de cena que de almuerzo?
- ¿La propina promedio también es mayor durante la cena?
- ¿El tamaño de las mesas cambia según el horario?

En este capítulo no vamos a responder todas esas preguntas. Pero sí podemos hacer una comparación adicional sencilla: revisar si la propina promedio también cambia entre almuerzo y cena.

```python
promedio_propina_almuerzo = df[df["time"] == "Lunch"]["tip"].mean()
promedio_propina_cena = df[df["time"] == "Dinner"]["tip"].mean()

comparacion_propinas = pd.DataFrame({
    "horario": ["Lunch", "Dinner"],
    "promedio_propina": [promedio_propina_almuerzo, promedio_propina_cena]
})

comparacion_propinas["promedio_propina"] = comparacion_propinas["promedio_propina"].round(2)

comparacion_propinas
```

La propina promedio también es mayor durante la cena.

En el almuerzo (`Lunch`), la propina promedio es de **2.73**. En la cena (`Dinner`), la propina promedio es de **3.10**.

Este resultado acompaña lo que habíamos visto con el total de la cuenta: en este dataset, tanto el promedio de cuenta como el promedio de propina son más altos durante la cena.

Este tipo de comparación empieza a mostrar una necesidad nueva: resumir información por grupos de manera más directa y ordenada.
