````python
import pandas as pd

datos = {
    "producto": [
        "Cuaderno", "Lapicera", "Mochila", "Regla", "Cartuchera",
        "Calculadora", "Auriculares", "Resaltadores", "Guardapolvo", "Compas",
        "Pendrive", "Carpeta", "Goma", "Lapiz", "Agenda"
    ],
    "categoria": [
        "Libreria", "Libreria", "Accesorios", "Libreria", "Accesorios",
        "Tecnologia", "Tecnologia", "Libreria", "Indumentaria", "Libreria",
        "Tecnologia", "Libreria", "Libreria", "Libreria", "Libreria"
    ],
    "precio": [
        1200, 350, 8500, 500, 3200,
        18500, 7600, 2100, 14500, 2600,
        9800, 950, 300, 250, 4300
    ],
    "cantidad_vendida": [
        15, 40, 5, 25, 8,
        3, 6, 12, 4, 10,
        7, 18, 30, 50, 9
    ],
    "sucursal": [
        "Centro", "Centro", "Norte", "Sur", "Norte",
        "Centro", "Sur", "Centro", "Norte", "Sur",
        "Centro", "Norte", "Sur", "Centro", "Sur"
    ],
    "forma_pago": [
        "Efectivo", "Debito", "Credito", "Efectivo", "Debito",
        "Credito", "Credito", "Debito", "Credito", "Efectivo",
        "Debito", "Efectivo", "Efectivo", "Debito", "Credito"
    ],
    "fecha": [
        "2024-03-01", "2024-03-01", "2024-03-02", "2024-03-02", "2024-03-03",
        "2024-03-03", "2024-03-04", "2024-03-04", "2024-03-05", "2024-03-05",
        "2024-03-06", "2024-03-06", "2024-03-07", "2024-03-07", "2024-03-08"
    ]
}

df = pd.DataFrame(datos)

df
````

Ese tipo de variedad permite que el ordenamiento sea útil. Si todos los precios o todas las cantidades fueran iguales, ordenar no aportaría demasiada información. En cambio, cuando los valores cambian de una fila a otra, ordenar ayuda a detectar rápidamente cuáles son los registros más altos, más bajos o más relevantes para una pregunta de análisis.

## Por qué ordenar datos

Ordenar datos significa reorganizar las filas de un `DataFrame` según los valores de una o más columnas.

Cuando miramos una tabla pequeña, tal vez podamos encontrar ciertos valores a simple vista. Pero a medida que el dataset crece, se vuelve mucho más difícil detectar cuáles son los precios más altos, cuáles son las cantidades más bajas, qué productos aparecen primero alfabéticamente o cómo se distribuyen los registros dentro de una categoría.

El ordenamiento nos ayuda a leer mejor la información. Por ejemplo, si ordenamos las ventas por `precio`, podemos identificar rápidamente los productos más baratos o más caros. Si ordenamos por `cantidad_vendida`, podemos ver qué productos tuvieron mayor o menor cantidad de unidades vendidas. Si ordenamos por `categoria`, podemos agrupar visualmente productos similares.

Ordenar no cambia el significado de los datos. Las filas siguen representando las mismas ventas, pero aparecen en otro orden. Esa reorganización puede facilitar mucho la interpretación.

En Pandas, la herramienta principal para ordenar filas según los valores de una columna es `sort_values()`.


## Ordenar por una columna

Para ordenar un `DataFrame` según los valores de una columna, usamos el método `sort_values()`.

Dentro de los paréntesis indicamos el nombre de la columna que queremos usar como criterio de ordenamiento.

Por ejemplo, si queremos ordenar las ventas según el precio, podemos escribir:

````python
df.sort_values("precio")
````

El resultado muestra el `DataFrame` ordenado por la columna `precio`.

Por defecto, Pandas ordena de menor a mayor. Por eso, al ordenar por `precio`, las primeras filas corresponden a los productos más baratos y las últimas a los productos más caros.

Esta instrucción no elimina filas ni columnas. Tampoco cambia los valores del dataset. Solamente reorganiza el orden en que aparecen las filas.

Es importante observar que el índice original se conserva. Aunque las filas aparezcan en otro orden, cada una mantiene el índice que tenía en el `DataFrame` original. Esto nos recuerda que ordenar modifica la forma en que vemos la tabla, pero no reinventa los registros.

## Orden ascendente y descendente

Cuando usamos `sort_values()`, Pandas ordena de manera ascendente por defecto.

En una columna numérica, ordenar de forma ascendente significa ir de menor a mayor. Por eso, cuando escribimos:

````python
df.sort_values("precio")
````

los productos aparecen ordenados desde el precio más bajo hasta el precio más alto.

Si queremos ordenar en sentido contrario, usamos el parámetro `ascending=False`.

````python
df.sort_values("precio", ascending=False)
````

Ahora el `DataFrame` aparece ordenado de mayor a menor precio. Las primeras filas corresponden a los productos más caros y las últimas a los productos más baratos.

El parámetro `ascending` controla el sentido del ordenamiento:

```text
ascending=True   → orden ascendente
ascending=False  → orden descendente
```

Como `ascending=True` es el valor por defecto, normalmente no hace falta escribirlo si queremos ordenar de menor a mayor. En cambio, sí debemos indicar `ascending=False` cuando queremos ordenar de mayor a menor.

Este detalle es muy útil cuando buscamos valores altos. Por ejemplo, si queremos ver rápidamente los productos más caros, conviene ordenar por precio en forma descendente.

## Ordenar por cantidad vendida

También podemos ordenar el `DataFrame` usando la columna `cantidad_vendida`.

Esto nos permite observar qué productos tuvieron menor o mayor cantidad de unidades vendidas.

Por ejemplo, podemos ordenar las ventas desde la menor cantidad vendida hasta la mayor:

````python
df.sort_values("cantidad_vendida")
````

En este caso, las primeras filas corresponden a los registros con menos unidades vendidas.

Si queremos ver primero los productos con mayor cantidad vendida, usamos `ascending=False`:

````python
df.sort_values("cantidad_vendida", ascending=False)
````

Ahora las primeras filas muestran las ventas con mayor cantidad de unidades.

Ordenar por cantidad vendida puede ayudarnos a detectar rápidamente qué productos se vendieron más y cuáles tuvieron menos movimiento. Sin embargo, todavía debemos interpretar estos resultados con cuidado: una cantidad vendida alta no siempre significa mayor ingreso total, porque también depende del precio de cada producto.

Por ejemplo, un producto barato puede vender muchas unidades, mientras que un producto caro puede vender pocas unidades y aun así representar un ingreso importante. Más adelante podremos crear nuevas columnas, como el importe total de cada venta, para analizar mejor ese tipo de situaciones.

## Ordenar columnas de texto

`sort_values()` también puede usarse con columnas de texto.

Cuando ordenamos una columna de texto, Pandas organiza los valores alfabéticamente. Por ejemplo, podemos ordenar el dataset según el nombre del producto.

````python
df.sort_values("producto")
````

El resultado muestra las filas ordenadas alfabéticamente por la columna `producto`.

También podemos ordenar por `categoria`:

````python
df.sort_values("categoria")
````

En este caso, los registros quedan agrupados visualmente por categoría. Primero aparecen las filas de una categoría, luego las de otra, y así sucesivamente, siguiendo el orden alfabético.

Ordenar columnas de texto puede ser útil para revisar datos categóricos, detectar valores escritos de manera diferente o simplemente presentar una tabla de forma más clara. Por ejemplo, si una columna de categorías tuviera valores como `"Libreria"`, `"libreria"` y `"Librería"`, al ordenar podríamos notar más fácilmente que existen variantes de escritura que tal vez deberían limpiarse más adelante.

También podemos ordenar textos de manera descendente:

````python
df.sort_values("producto", ascending=False)
````

En este caso, el orden alfabético se invierte.

Aunque muchas veces usamos `sort_values()` con columnas numéricas, también es una herramienta útil para explorar columnas categóricas o textuales.

## Ordenar por más de una columna

A veces una sola columna no alcanza para organizar bien la información.

Por ejemplo, si ordenamos solamente por `categoria`, los productos quedan agrupados por categoría, pero dentro de cada categoría pueden aparecer en cualquier orden. Tal vez queremos que, dentro de cada categoría, los productos aparezcan ordenados por precio.

Para eso podemos ordenar usando más de una columna.

En `sort_values()`, en lugar de pasar un solo nombre de columna, pasamos una lista de columnas.

````python
df.sort_values(["categoria", "precio"])
````

Esta instrucción ordena primero por `categoria` y, dentro de cada categoría, ordena por `precio`.

Podemos leerlo así: *agrupá visualmente los registros por categoría y, dentro de cada grupo, mostrá los productos desde el precio más bajo hasta el precio más alto*.

También podemos ordenar por `sucursal` y luego por `cantidad_vendida`:

````python
df.sort_values(["sucursal", "cantidad_vendida"])
````

En este caso, las filas se ordenan primero por `sucursal`. Luego, dentro de cada sucursal, se ordenan por `cantidad_vendida`.

Ordenar por más de una columna es útil cuando queremos organizar la información con distintos niveles de criterio. Primero definimos el criterio principal y después el criterio secundario.

## Ordenar varias columnas con distintos criterios

Cuando ordenamos por más de una columna, también podemos indicar si cada una debe ordenarse de forma ascendente o descendente.

Para eso usamos el parámetro `ascending`, pero en lugar de pasar un solo valor, pasamos una lista de valores.

Por ejemplo, podemos ordenar por `categoria` de manera ascendente y, dentro de cada categoría, ordenar por `precio` de manera descendente.

````python
df.sort_values(
    ["categoria", "precio"],
    ascending=[True, False]
)
````

En esta instrucción, la primera columna de ordenamiento es `categoria`, y se ordena alfabéticamente porque indicamos `True`. La segunda columna es `precio`, y se ordena de mayor a menor dentro de cada categoría porque indicamos `False`.

Esto permite construir ordenamientos más expresivos. Por ejemplo, si queremos revisar las ventas por sucursal y, dentro de cada sucursal, ver primero las mayores cantidades vendidas, podemos escribir:

````python
df.sort_values(
    ["sucursal", "cantidad_vendida"],
    ascending=[True, False]
)
````

En este caso, Pandas ordena primero por `sucursal` y luego, dentro de cada sucursal, muestra las ventas con mayor cantidad vendida al principio.

Cuando usamos varias columnas para ordenar, el orden de las listas importa. La primera columna es el criterio principal, la segunda columna es el criterio secundario, y así sucesivamente.

También debe coincidir la cantidad de columnas con la cantidad de valores en `ascending`. Si ordenamos por dos columnas, debemos indicar dos valores: uno para cada criterio.

## Combinar filtros y ordenamiento

En muchos análisis no ordenamos todo el dataset, sino un subconjunto de filas.

Por ejemplo, podemos querer ver solamente los productos de la categoría `"Tecnologia"` y ordenarlos por precio de mayor a menor.

Para hacerlo, primero filtramos y luego aplicamos `sort_values()` al resultado.

````python
df[df["categoria"] == "Tecnologia"].sort_values("precio", ascending=False)
````

La primera parte de la instrucción filtra el `DataFrame`:

```python
df[df["categoria"] == "Tecnologia"]
```

Luego, sobre ese resultado filtrado, aplicamos:

```python
.sort_values("precio", ascending=False)
```

De esta manera obtenemos únicamente las ventas de la categoría `"Tecnologia"`, ordenadas desde el precio más alto hasta el precio más bajo.

También podemos combinar un filtro más específico con ordenamiento. Por ejemplo, podemos seleccionar las ventas realizadas en la sucursal `"Centro"` y ordenarlas por `cantidad_vendida` de mayor a menor.

````python
df[df["sucursal"] == "Centro"].sort_values("cantidad_vendida", ascending=False)
````

Este encadenamiento es muy común en Pandas. Primero reducimos el dataset con un filtro y luego organizamos el resultado para leerlo mejor.

La estructura general es:

```python
df[condición].sort_values("columna")
```

o, si queremos ordenar de mayor a menor:

```python
df[condición].sort_values("columna", ascending=False)
```

Esta forma de trabajo permite responder preguntas más precisas. No solo preguntamos qué filas cumplen una condición, sino también cómo se ordenan esas filas según una variable relevante.

## Seleccionar columnas después de ordenar

Cuando ordenamos un `DataFrame`, Pandas muestra todas las columnas. Sin embargo, muchas veces nos interesa presentar solamente algunas.

Por ejemplo, si queremos ver los productos más caros, tal vez no necesitemos mostrar la fecha o la forma de pago. Podría alcanzarnos con ver el producto, la categoría y el precio.

Podemos ordenar primero y luego seleccionar algunas columnas:

````python
df.sort_values("precio", ascending=False)[["producto", "categoria", "precio"]]
````

Esta instrucción se lee de izquierda a derecha. Primero ordenamos el `DataFrame` por `precio` de mayor a menor. Luego, del resultado ordenado, seleccionamos solamente las columnas `producto`, `categoria` y `precio`.

También podemos hacer algo parecido con la cantidad vendida:

````python
df.sort_values("cantidad_vendida", ascending=False)[
    ["producto", "categoria", "cantidad_vendida"]
]
````

En este caso, la tabla resultante permite identificar rápidamente qué productos tuvieron mayor cantidad de unidades vendidas.

Esta forma de trabajo es útil cuando queremos construir salidas más limpias. En lugar de mostrar toda la tabla, podemos quedarnos con las columnas necesarias para responder una pregunta concreta.

También podríamos combinar filtro, ordenamiento y selección de columnas:

````python
df[df["categoria"] == "Libreria"].sort_values(
    "cantidad_vendida",
    ascending=False
)[["producto", "categoria", "cantidad_vendida", "sucursal"]]
````

Esta instrucción selecciona solamente las ventas de la categoría `"Libreria"`, las ordena por `cantidad_vendida` de mayor a menor y muestra algunas columnas relevantes.

Cuando las operaciones se encadenan, el orden importa. En este ejemplo, primero filtramos, después ordenamos y finalmente seleccionamos columnas para presentar el resultado de manera más clara.

## Guardar un resultado ordenado

Cuando usamos `sort_values()`, Pandas muestra el `DataFrame` ordenado, pero no modifica el `DataFrame` original.

Por ejemplo, podemos ordenar por precio de mayor a menor:

````python
df.sort_values("precio", ascending=False)
````

Si después mostramos nuevamente `df`, veremos que el orden original se conserva.

````python
df
````

Esto ocurre porque `sort_values()` devuelve un nuevo resultado ordenado, pero no cambia automáticamente el `DataFrame` sobre el que se aplicó.

Si queremos reutilizar el resultado ordenado más adelante, podemos guardarlo en una nueva variable.

````python
ventas_por_precio = df.sort_values("precio", ascending=False)

ventas_por_precio
````

Ahora `ventas_por_precio` contiene el dataset ordenado por precio de mayor a menor, mientras que `df` sigue conservando el orden original.

Esta forma de trabajo es recomendable porque evita modificar accidentalmente los datos originales. Podemos tener distintas versiones ordenadas del mismo dataset, según la pregunta que queramos responder.

Por ejemplo:

````python
ventas_por_cantidad = df.sort_values("cantidad_vendida", ascending=False)

ventas_por_cantidad
````

En este caso, la variable `ventas_por_cantidad` contiene las ventas ordenadas según la cantidad vendida, de mayor a menor.

Guardar resultados en variables con nombres claros ayuda a que el análisis sea más legible. Un nombre como `ventas_por_precio` o `ventas_por_cantidad` permite entender rápidamente qué contiene cada objeto.

## Ordenar por índice

Hasta ahora ordenamos el `DataFrame` usando los valores de una columna. Para eso usamos `sort_values()`.

Pero Pandas también permite ordenar un `DataFrame` según su índice. Para eso usamos `sort_index()`.

Primero veamos un ejemplo. Podemos crear un `DataFrame` ordenado por precio de mayor a menor:

````python
ventas_por_precio = df.sort_values("precio", ascending=False)

ventas_por_precio
````

Observemos que las filas aparecen ordenadas por precio, pero el índice conserva las etiquetas originales. Por eso, al mirar la columna del índice, los valores pueden aparecer en un orden como `5`, `8`, `10`, `2`, etc.

Si queremos ordenar nuevamente ese resultado según el índice, podemos usar `sort_index()`.

````python
ventas_por_precio.sort_index()
````

Ahora las filas vuelven a aparecer ordenadas según las etiquetas del índice.

La diferencia principal es esta:

```text
sort_values() → ordena según los valores de una o más columnas
sort_index()  → ordena según el índice
```

En este punto no necesitamos profundizar demasiado en el índice. Más adelante lo vamos a estudiar con más detalle. Por ahora, alcanza con reconocer que Pandas puede ordenar tanto por valores de columnas como por etiquetas del índice.

## Errores frecuentes al ordenar datos

Al ordenar datos con Pandas, pueden aparecer algunos errores o confusiones frecuentes.

Uno de los errores más comunes es escribir mal el nombre de la columna. Por ejemplo, nuestro `DataFrame` tiene una columna llamada `precio`, con minúscula. Si intentamos ordenar usando `"Precio"` con mayúscula inicial, Pandas no podrá encontrar esa columna.

````python
# Esta instrucción produciría un error porque la columna se llama "precio", no "Precio".
# df.sort_values("Precio")
````

Para evitar este problema, conviene revisar los nombres de las columnas antes de ordenar.

````python
df.columns
````

Otro error frecuente es olvidar `ascending=False` cuando queremos ordenar de mayor a menor. Si escribimos:

```python
df.sort_values("precio")
```

Pandas ordena de menor a mayor, porque ese es el comportamiento por defecto. Si queremos ver primero los precios más altos, debemos escribir:

```python
df.sort_values("precio", ascending=False)
```

También puede ocurrir que pensemos que `sort_values()` modifica el `DataFrame` original. Sin embargo, como vimos antes, el método devuelve un resultado ordenado, pero no cambia automáticamente `df`.

Por eso, si queremos conservar el resultado ordenado, conviene guardarlo en una nueva variable:

```python
ventas_ordenadas = df.sort_values("precio", ascending=False)
```

Finalmente, cuando ordenamos por varias columnas, debemos cuidar que la lista de columnas y la lista de valores de ascending tengan la misma longitud.

```python
# Esta instrucción produciría un error porque se indican dos columnas,
# pero solo un valor para ascending dentro de una lista.

# df.sort_values(["categoria", "precio"], ascending=[True])
```

La forma correcta sería:

```python
df.sort_values(["categoria", "precio"], ascending=[True, False])
```

En general, para ordenar datos de manera segura conviene revisar tres cosas: que el nombre de la columna exista, que el sentido del ordenamiento sea el que necesitamos y que sepamos si queremos solamente ver el resultado ordenado o guardarlo para usarlo más adelante.

Estas operaciones nos permitirán responder preguntas como cuáles son los productos más caros, cuáles son los más baratos, cuáles tienen mayor cantidad vendida o cuáles aparecen al final de un ordenamiento determinado.
