# Capítulo 039 — Escribir conclusiones a partir de los datos

## Pasar del resultado a la conclusión

En el capítulo anterior integramos varias herramientas de análisis exploratorio. Revisamos la estructura del dataset, resumimos variables, contamos categorías, construimos gráficos, comparamos grupos y observamos relaciones entre variables.

Ahora vamos a trabajar sobre una parte distinta del análisis: la escritura de conclusiones.

Una conclusión no es simplemente copiar un número o describir un gráfico. Es una frase que interpreta un resultado, lo conecta con una pregunta y reconoce hasta dónde podemos llegar con los datos disponibles.

Por ejemplo, no es lo mismo escribir:

`Dinner tiene un promedio de 20.80`

que escribir:

`En este dataset, las cuentas de Dinner tienden a tener importes más altos que las de Lunch, aunque la comparación debe leerse considerando que Dinner tiene muchos más registros.`

La primera frase solo informa un valor. La segunda convierte ese valor en una conclusión más útil, porque lo compara, lo interpreta y marca una precaución.

## Qué vamos a trabajar en este capítulo

En este capítulo vamos a usar algunos resultados obtenidos en el análisis exploratorio anterior, pero el foco no estará en volver a calcularlos desde cero. El objetivo será aprender a redactar conclusiones a partir de esos resultados.

Vamos a distinguir entre:

- observaciones;
- interpretaciones;
- conclusiones;
- límites del análisis.

También vamos a practicar una idea importante: una buena conclusión debe decir algo relevante, pero no debe afirmar más de lo que los datos permiten sostener.

Para eso vamos a trabajar con ejemplos basados en el dataset `tips`.

## Resultados de partida

Tomaremos como punto de partida algunos resultados ya obtenidos:

- el dataset tiene 244 cuentas registradas y no presenta valores faltantes;
- la mayoría de las cuentas se concentra en importes bajos o medios;
- las propinas también se concentran en valores bajos o medios;

- la correlación entre `total_bill` y `tip` es aproximadamente 0.68.

A partir de estos resultados vamos a construir frases de análisis más claras y cuidadas.

## Observación, interpretación y conclusión

Antes de escribir conclusiones, conviene separar tres niveles de escritura.

Una **observación** describe algo que aparece directamente en una tabla o gráfico. Por ejemplo: `Dinner tiene 176 registros y Lunch tiene 68`.

Una **interpretación** explica qué puede significar esa observación dentro del análisis. Por ejemplo: `la comparación entre horarios no está basada en grupos del mismo tamaño`.

Una **conclusión** integra la observación y la interpretación para responder una pregunta. Por ejemplo: `En este dataset, Dinner concentra más registros y presenta importes promedio más altos que Lunch, pero la diferencia debe leerse considerando el tamaño desigual de los grupos`.

Esta separación ayuda a escribir con más precisión. No todas las frases de un análisis tienen que ser conclusiones finales; muchas veces primero necesitamos describir, luego interpretar y recién después concluir.

## Mejorar frases de análisis

Una misma información puede escribirse de maneras muy distintas. Algunas frases son demasiado pobres porque solo repiten un número. Otras son problemáticas porque exageran lo que los datos permiten afirmar.

La escritura de conclusiones requiere equilibrio: necesitamos decir algo claro, pero también respetar los límites del dataset.

Veamos algunos ejemplos.

## Ejemplos de mejora de frases

### Ejemplo 1: cantidad de registros por horario

**Resultado observado:**  
`Dinner` tiene 176 registros y `Lunch` tiene 68.

**Frase débil:**  
Hay más cenas que almuerzos.

**Frase mejorada:**  
En este dataset, `Dinner` concentra muchos más registros que `Lunch`, por lo que las comparaciones entre horarios deben leerse considerando que los grupos tienen tamaños diferentes.

### Ejemplo 2: promedio de cuenta por horario

**Resultado observado:**  
El promedio de `total_bill` es 20.80 en `Dinner` y 17.17 en `Lunch`.

**Frase débil:**  
Dinner es más caro.

**Frase mejorada:**  
En este dataset, las cuentas de `Dinner` tienen un importe promedio más alto que las de `Lunch`. Sin embargo, esto describe los datos disponibles y no permite afirmar que todas las cenas sean más caras que todos los almuerzos.

### Ejemplo 3: pocos registros en un grupo

**Resultado observado:**  
`Fri` tiene solo 19 registros.

**Frase débil:**  
Friday no sirve.

**Frase mejorada:**  
`Fri` tiene pocos registros en comparación con los demás días, por lo que sus promedios y comparaciones deben interpretarse con más cautela.

### Ejemplo 4: correlación entre variables

**Resultado observado:**  
La correlación entre `total_bill` y `tip` es 0.68.

**Frase débil:**  
El total causa la propina.

**Frase mejorada:**  
Existe una relación positiva moderada entre `total_bill` y `tip`: en general, las cuentas más altas tienden a asociarse con propinas más altas. Esto no implica causalidad ni explica por sí solo el comportamiento de las propinas.

La tabla muestra que una conclusión no debería limitarse a repetir un resultado numérico. También debe explicar qué significa ese resultado dentro del análisis.

Por ejemplo, decir que `Dinner` tiene 176 registros y `Lunch` tiene 68 es correcto, pero todavía no alcanza. La conclusión mejora cuando conectamos ese dato con una consecuencia analítica: las comparaciones entre horarios no están sostenidas por la misma cantidad de observaciones.

También aparece otro cuidado importante: evitar afirmaciones demasiado generales. Si en este dataset el promedio de `Dinner` es más alto que el de `Lunch`, podemos decir eso mismo, pero no deberíamos convertirlo en una regla universal sobre todos los restaurantes, todas las cenas o todos los almuerzos.

Una buena conclusión suele combinar tres elementos: el resultado observado, una interpretación razonable y una advertencia sobre el alcance de esa interpretación.

## Evitar conclusiones exageradas

Uno de los errores más comunes al escribir conclusiones es exagerar el resultado. Esto puede ocurrir cuando usamos palabras demasiado fuertes, cuando generalizamos más allá del dataset o cuando confundimos una asociación con una causa.

Por ejemplo, a partir del análisis de `tips`, sería exagerado escribir:

`Las personas siempre dejan más propina cuando la cuenta es más alta.`

El gráfico de dispersión y la correlación sugieren una relación positiva entre `total_bill` y `tip`, pero no muestran una regla perfecta. Hay cuentas con importes parecidos y propinas diferentes. Además, el dataset no contiene toda la información necesaria para explicar por qué una persona dejó una propina determinada.

Una versión más prudente sería:

`En este dataset, las cuentas con importes más altos tienden a estar asociadas con propinas más altas, aunque la relación no es perfecta y no permite afirmar causalidad.`

## Algunas expresiones útiles

Al escribir conclusiones exploratorias, pueden ser útiles algunas expresiones que ayudan a mantener la prudencia.

**En este dataset...**  
Sirve para recordar que la conclusión se refiere a los datos analizados, no a una regla general.

Ejemplo:  
`En este dataset, Dinner concentra más registros que Lunch.`

**Los datos sugieren...**  
Sirve para presentar una lectura posible sin convertirla en una afirmación definitiva.

Ejemplo:  
`Los datos sugieren una relación positiva entre total_bill y tip.`

**Tiende a...**  
Sirve para describir patrones generales, especialmente cuando no todos los casos siguen exactamente la misma regla.

Ejemplo:  
`La propina tiende a aumentar cuando aumenta el importe total de la cuenta.`

**Debe interpretarse con cuidado...**  
Sirve para señalar límites, como pocos registros o grupos desbalanceados.

Ejemplo:  
`El promedio de Fri debe interpretarse con cuidado porque hay pocos registros.`

**No alcanza para afirmar...**  
Sirve para evitar conclusiones causales o demasiado fuertes.

Ejemplo:  
`Esta relación no alcanza para afirmar causalidad.`

**Dentro de los datos analizados...**  
Funciona de manera parecida a “en este dataset” y ayuda a ubicar el alcance de la conclusión.

Ejemplo:  
`Dentro de los datos analizados, Sun presenta el mayor promedio de cuenta.`

Estas expresiones no son fórmulas obligatorias, pero ayudan a construir un tono analítico más preciso. En análisis exploratorio, muchas veces no buscamos demostrar una afirmación definitiva, sino describir patrones, diferencias y relaciones que aparecen en los datos.

La clave está en evitar dos extremos. Por un lado, no queremos escribir frases tan débiles que solo repitan números sin interpretación. Por otro lado, tampoco queremos conclusiones tan fuertes que los datos no puedan sostener.

El objetivo es escribir frases claras, útiles y honestas con la información disponible.

## Describir, comparar y concluir

Cuando escribimos sobre datos, no todas las frases cumplen la misma función. Algunas frases describen un resultado, otras comparan valores y otras construyen una conclusión más completa.

Una **descripción** dice qué aparece en una tabla o gráfico. Por ejemplo: `Sun tiene un promedio de cuenta de 21.41`.

Una **comparación** pone ese valor en relación con otros. Por ejemplo: `Sun tiene un promedio más alto que Thur, Fri y Sat`.

Una **conclusión** interpreta esa comparación dentro del análisis. Por ejemplo: `Dentro de los datos analizados, Sun presenta el mayor importe promedio de cuenta, aunque esta lectura debe considerarse junto con la cantidad de registros disponibles para cada día`.

Las tres formas son útiles. El problema aparece cuando un análisis se queda solo en la descripción y no avanza hacia una interpretación, o cuando salta demasiado rápido a una conclusión sin mostrar en qué resultado se apoya.

## Ejemplo con el promedio por día

Tomemos uno de los resultados del análisis anterior: los importes promedio de cuenta por día.

Los valores eran:

- `Thur`: 17.68
- `Fri`: 17.15
- `Sat`: 20.44
- `Sun`: 21.41

Una descripción posible sería:

`Sun tiene un importe promedio de cuenta de 21.41.`

Una comparación agregaría más contexto:

`Sun tiene el importe promedio de cuenta más alto, seguido por Sat.`

Una conclusión más completa podría ser:

`Dentro de los datos analizados, Sun y Sat presentan importes promedio de cuenta más altos que Thur y Fri. Esta diferencia sugiere un cambio entre los días registrados de fin de semana y los otros días, aunque la interpretación debe considerar que la cantidad de registros no es igual para todos los días.`

En el ejemplo anterior, la conclusión no agrega una técnica nueva. Lo que hace es organizar mejor la lectura del resultado.

Primero identifica el patrón principal: `Sun` y `Sat` tienen promedios más altos. Después evita una afirmación demasiado fuerte: no dice que “los fines de semana siempre se gasta más”, porque el dataset no permite sostener una conclusión general de ese tipo. Finalmente, incorpora una precaución metodológica: los días no tienen la misma cantidad de registros.

Esa combinación es una buena práctica al escribir conclusiones exploratorias: decir algo relevante, pero mantener claro el alcance de la afirmación.

## Usar límites como parte de la conclusión

A veces se piensa que mencionar límites debilita una conclusión. En realidad, suele mejorarla. Una conclusión con límites claros es más confiable que una afirmación demasiado amplia.

Por ejemplo, podríamos escribir:

`Dinner tiene cuentas más altas que Lunch.`

Esa frase apunta en una dirección correcta, pero queda incompleta. Una versión mejor sería:

`En este dataset, Dinner presenta un importe promedio de cuenta más alto que Lunch. Sin embargo, Dinner también tiene muchos más registros, por lo que la comparación debe leerse considerando el tamaño desigual de los grupos.`

La segunda versión no niega el resultado. Lo hace más preciso.

## Tipos de límites frecuentes

En análisis exploratorio, algunos límites aparecen con mucha frecuencia. Conviene aprender a reconocerlos para incorporarlos a la escritura cuando corresponda.

Un primer límite es el **tamaño del dataset**. Si tenemos pocos registros, las conclusiones deben ser más prudentes.

Otro límite es el **desbalance entre grupos**. Si un grupo tiene muchos datos y otro tiene muy pocos, comparar promedios puede ser útil, pero requiere cuidado.

También está el límite del **contexto disponible**. Si no sabemos cómo se recolectaron los datos, en qué período, en qué lugar o bajo qué condiciones, no conviene generalizar demasiado.

Finalmente, existe el límite de la **causalidad**. Que dos variables se relacionen no significa automáticamente que una cause la otra.

## Convertir hallazgos en un párrafo

Hasta ahora vimos frases individuales. Pero en un informe o en un cuaderno de análisis, muchas veces necesitamos escribir párrafos completos.

Un buen párrafo de conclusión no debería ser una lista desordenada de resultados. Conviene que tenga una idea principal y que los resultados funcionen como apoyo de esa idea.

Por ejemplo, si queremos escribir sobre diferencias entre días, no alcanza con acumular datos sueltos. Necesitamos conectar los valores, compararlos y marcar alguna precaución.

Supongamos que queremos escribir una conclusión sobre los importes por día.

Una versión débil podría ser:

`Thur tiene 17.68. Fri tiene 17.15. Sat tiene 20.44. Sun tiene 21.41. Sun es el mayor.`

Esa frase no es incorrecta, pero se parece más a una repetición de la tabla que a una conclusión.

Una versión más clara podría ser:

`Dentro de los datos analizados, los días con mayor importe promedio de cuenta son Sun y Sat. Ambos superan a Thur y Fri, lo que sugiere una diferencia entre los días registrados de fin de semana y los demás días del dataset. Sin embargo, esta comparación debe leerse con cautela porque la cantidad de registros no es igual para todos los días, especialmente en Fri, que tiene pocos casos.`

La segunda versión organiza mejor la lectura. No repite todos los valores, identifica el patrón principal y agrega un límite importante para interpretar la comparación.

## Una estructura posible para escribir conclusiones

Una forma simple de escribir un párrafo de conclusión es seguir esta estructura:

1. Presentar el patrón principal.
2. Mencionar el resultado que lo sostiene.
3. Agregar una interpretación prudente.
4. Señalar un límite si corresponde.

Por ejemplo:

`En este dataset, Dinner presenta importes de cuenta más altos que Lunch. Esta lectura se apoya en que Dinner tiene un promedio de total_bill de 20.80, frente a 17.17 en Lunch. La diferencia sugiere que las cuentas de cena tienden a ser más altas dentro de los datos analizados. Sin embargo, Dinner también tiene muchos más registros que Lunch, por lo que la comparación debe interpretarse considerando el tamaño desigual de los grupos.`

Esta estructura no es una fórmula obligatoria, pero ayuda a evitar dos problemas frecuentes: quedarse solo en números sueltos o escribir conclusiones demasiado generales.

## Conclusiones sobre relaciones entre variables

Cuando escribimos sobre relaciones entre variables numéricas, también debemos ser cuidadosos. En el análisis anterior vimos que la correlación entre `total_bill` y `tip` era aproximadamente 0.68, y que el gráfico de dispersión mostraba una tendencia ascendente.

Una conclusión débil sería:

`total_bill y tip están relacionados.`

Una conclusión exagerada sería:

`Si aumenta total_bill, aumenta automáticamente tip.`

Una versión más adecuada sería:

`En este dataset, total_bill y tip presentan una relación positiva moderada. Las cuentas con importes más altos tienden a estar asociadas con propinas más altas, aunque la relación no es perfecta y no permite afirmar causalidad.`

Esta última frase es mejor porque indica la dirección de la relación, reconoce que hay variabilidad y evita transformar una correlación en una causa.

## Evitar tres errores comunes

Al redactar conclusiones a partir de datos, conviene evitar tres errores frecuentes.

El primero es **copiar números sin interpretarlos**. Los valores son importantes, pero una conclusión debe explicar qué significan dentro del análisis.

El segundo es **generalizar más allá del dataset**. Si los datos corresponden a una muestra específica, no podemos convertirlos automáticamente en una regla universal.

El tercero es **confundir relación con causalidad**. Una correlación o una tendencia visual puede mostrar asociación, pero no demuestra por sí sola que una variable cause a la otra.

Evitar estos errores ayuda a escribir conclusiones más claras, más útiles y más honestas con los datos disponibles.

## De varios hallazgos a un cierre general

Al final de un análisis, podemos reunir varios hallazgos en un cierre general. Ese cierre no debería intentar repetir todo lo que se hizo, sino seleccionar las ideas más importantes.

Por ejemplo, a partir del análisis exploratorio del dataset `tips`, podríamos escribir un cierre como este:

`El análisis exploratorio muestra que la mayoría de las cuentas y propinas se concentra en valores bajos o medios, con algunos valores altos menos frecuentes. También se observan diferencias entre días y horarios: Sat y Sun presentan importes promedio de cuenta más altos, mientras que Dinner concentra más registros y tiene un promedio mayor que Lunch. Además, total_bill y tip muestran una relación positiva moderada. Estos hallazgos deben interpretarse dentro del alcance del dataset, considerando que algunos grupos tienen pocos registros y que no contamos con información contextual adicional sobre la recolección de los datos.`

Este párrafo integra distribución, comparación entre grupos, relación entre variables y límites del análisis.

## Cierre del capítulo

En este capítulo trabajamos sobre la escritura de conclusiones a partir de datos. El foco no estuvo en calcular nuevos resultados, sino en aprender a comunicar mejor los resultados de un análisis exploratorio.

Vimos que una buena conclusión no se limita a copiar números ni a describir un gráfico. Una conclusión útil conecta un resultado con una interpretación, responde una pregunta y reconoce los límites de lo que puede afirmarse.

También diferenciamos entre observación, interpretación y conclusión. Esta separación ayuda a escribir con más precisión: primero identificamos qué muestran los datos, luego pensamos qué significa ese resultado dentro del análisis y finalmente redactamos una conclusión prudente.

Además, revisamos errores frecuentes, como generalizar más allá del dataset, usar frases demasiado fuertes o confundir correlación con causalidad. En análisis exploratorio, muchas veces trabajamos con patrones y señales, no con demostraciones definitivas.

La idea central del capítulo es que escribir conclusiones también forma parte del análisis de datos. No alcanza con producir tablas y gráficos: necesitamos convertir esos resultados en explicaciones claras, honestas y útiles para quien lee.

En el próximo capítulo vamos a usar esta idea para armar un informe breve de análisis exploratorio, organizando preguntas, resultados, visualizaciones y conclusiones dentro de una estructura más completa.
